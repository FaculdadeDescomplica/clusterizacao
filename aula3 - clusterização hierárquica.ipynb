## Abordagem Aglomerativa vs. Abordagem Divisiva

### Abordagem Aglomerativa

A estratégia Aglomerativa, em inglês Agglomerative Clustering, é uma forma de resolver problemas de agrupamento em que,de partida, todos as observações da base de dados são tratados como um cluster, ou grupo, individual. Em seguida, se calcula a distância de todos as observações umas para as outras e, por fim, os pontos mais próximos vão sendo combinados até que exista um grupo que contenha todas as observações.

**Algoritmo**:
* Iniciar tratando cada observação como um cluster
* Definir o melhor par de clusters para combinar
* Combinar o par de clusters escolhido
* Repetir até que todas as observações estejam em um único cluster



### Abordagem Divisiva

A estratégia Divisiva, em inglês Divisive Clustering, é uma abordagem para resolver problemas de agrupamento que, de partida, considera que todas as observações fazem parte de um único grupo. Em seguida, utilizando algum método para dividir a base de dados cria grupos e subgrupos até que todas as observações tenham formado grupos individuais.

**Algoritmo**:
* Iniciar todas as observações em um único cluster
* Sub-dividir o cluster em dois novos clusters
* Em cada um dos novos clusters aplicar essa estratégia e sub-dividi-los até que cada observação seja um cluster

---

Uma vantagem do Agrupamento Hierárquico que funciona em ambas as abordagens é que não é necessário definir o número de grupos a serem encontrados a priori.


## Dendrograma

A estrutura de árvore que surge ao se utilizar os métodos de Agrupamento Hierárquico é chamada de **dendrograma**. É um diagrama que nos ajuda a visualizar e entender as distâncias e relações entre as observações da nossa base. É uma representação visual das Matriz de Distâncias.




O Dendrograma pode ser utilizado para definir quandos grupos existem nos dados que estamos analisando.



````python
import pandas as pd
import seaborn as sns
sns.set_theme(style="darkgrid", palette="pastel")
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
````

````python
df = pd.read_csv('data/paises_do_mundo.csv')
df.head()
````

````python
paises = ['Angola', 'Argentina', 'Australia', 'Brazil', 'Chile',
          'Japan', 'Morocco', 'Senegal', 'South Korea']

mini_df = df[df.pais.str.strip().isin(paises)].copy()
X = mini_df.iloc[:, 4:10].dropna().copy()
````

````python
Z = linkage(X, method='single')

plt.figure(figsize=(7,5))
dendrogram(Z, labels=list(mini_df.pais),
           orientation='top',
           leaf_font_size=10,leaf_rotation=60,);
````

## Métodos Hierárquicos

Quando trabalhamos com a Matrix de Distâncias entre as observações é relatimente simples entender como a distância entre duas observações é calculada, porém, quando precisamos calcular a distância entre grupos essa tarefa pode ficar um pouco mais desafiadora. Existem várias estratégias para calcular a **Distância Inter-Cluster** ou **Dissimilaridade Inter-Cluster**. Aqui vamos listar alguns dos mais utilizados.

*Linkage pode ser entendido como conexão*


* **Single Linkage:** A distância entre dois grupos é definida pela distância entre os dois membros de cada grupo que estão mais **próximos**.


* **Complete Linkage:** A distância entre dois grupos é definida pela distância entre os dois membros de cada grupo que estão mais **distantes**.


* **Average Linkage:** A distância entre dois grupos é definida pela **média** das distâncias de todos os membros de um grupo para todos os membros do outro grupo.


* **Ward's Linkage:**
    * Para cada grupo é definido calculado o valor central (média)
    * Em cada grupo a distância de todos os membros para o valor central é somada
    * Combina os grupos que ao serem unidos geram a menor soma das distâncias intra-cluster
    
    

## Clusterização Hierárquica em Python

Utilizamos a bliblioteca **scipy** para trabalhar com algoritmos de agrupamento. Aqui vamos utilizar trabalhar com os módulos ```dendrogram``` para visualização e ```linkage``` para fazer realizar o agrupamento.

Para utilizar o **linkage** temos três parametros principais:
* y: a base que sera agrupada
* method: a forma de calcular a distância inter-clusters
* metric: qual será a métrica de distância utilizada

````python
import pandas as pd
import seaborn as sns
sns.set_theme(style="darkgrid", palette="pastel")
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
````

````python
df = pd.read_csv('data/paises_do_mundo.csv')
df.head()
````

````python
df[df.pais.str.strip().str.contains('K')]
````

````python
paises = ['Angola', 'Argentina', 'Australia', 'Brazil', 'Chile',
          'Japan', 'Morocco', 'Senegal', 'Korea, South']

mini_df = df[df.pais.str.strip().isin(paises)].copy()

X = mini_df.iloc[:, 4:10].dropna().copy()
````

````python
Z = linkage(y=X, method='ward')

plt.figure(figsize=(7,5))
dendrogram(Z, labels=list(mini_df.pais),
           orientation='top',
           leaf_font_size=10,leaf_rotation=60,);
````

````python
Z = linkage(X, method='single')

plt.figure(figsize=(7,5))
dendrogram(Z, labels=list(mini_df.pais),
           orientation='top',
           leaf_font_size=10,leaf_rotation=60,);
````

````python
Z = linkage(X, method='centroid')

plt.figure(figsize=(7,5))
dendrogram(Z, labels=list(mini_df.pais),
           orientation='top',
           leaf_font_size=10,leaf_rotation=60,);
````

## Clusterização Hierárquica na Prática

````python
import pandas as pd
import numpy as np

import seaborn as sns
sns.set_theme(style="darkgrid", palette="pastel")
from matplotlib import pyplot as plt

from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
````

````python
data = pd.read_csv('data/tamanho_pinguins.csv')
data.head()
````

````python
# nesse caso iremos apenas apagar as linhas que possuem valores faltantes e outliers
data = data.dropna()
data = data[data.larg_bico_mm < 50]
````

````python
data.shape
````

````python
X, y = data.iloc[:, 2:-1], data.iloc[:, 0]

scaler = MinMaxScaler()
X_ = scaler.fit_transform(X)
X_
````

````python
x = X_[:, [0,2]]
kmeans = KMeans(n_clusters=3, random_state=3, n_init="auto").fit(x)
labels = kmeans.labels_

plt.scatter(x[labels == 0, 0], x[labels == 0, 1], color="green", edgecolors='black', s=15)
plt.scatter(x[labels == 1, 0], x[labels == 1, 1], color="red", edgecolors='black', s=15)
plt.scatter(x[labels == 2, 0], x[labels == 2, 1], color="blue", edgecolors='black', s=15)
plt.xlabel('comp_bico_mm')
plt.ylabel('comp_asa_mm');
````

````python
df_res = pd.DataFrame(x, columns = ['comp_bico_mm', 'comp_asa_mm'])
df_res['cluster'] = labels
````

````python
def k_means(data, k, threshhold=2):

    n = np.size(data, 0)
    
    # atribui os pontos a grupos aleatorios
    cluster_centers = np.random.choice(range(0, n), k)
    clustering = np.random.randint(0, k, n)
    cluster_means = data[cluster_centers]
    old_clustering = np.zeros(n)
    
    # enquanto a clusterização não converge ... 
    while np.sum(clustering != old_clustering) > threshhold:
        old_clustering = clustering

        # passo 1: Atribui pontos para os grupos
        cluster_distances = np.zeros((n, k))
        for cluster in range(k):
            cluster_distances[:, cluster] = np.sum(np.sqrt((data - cluster_means[cluster])**2), 1)
        clustering = np.argmin(cluster_distances, 1)

        # step 2: recalcula as médias
        # cluster_means deve ser um array k x d
        cluster_means = np.array([np.mean(data[clustering==c],0) for c in range(k)])
    return cluster_means, clustering
````
